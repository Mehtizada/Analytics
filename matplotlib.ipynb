```python
data = np.linspace (0, 10, 100)
data
```

```python
plt.plot(data, np.exp(data));
```

```python
plt.scatter(data, np.exp(data));
```

```python
plt.plot(data, np.sin(data));
```

```python
fig,ax = plt.subplots (figsize = (10,5))
ax.plot(data, np.sin(data))
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title')
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots (nrows = 2 , ncols = 2 , figsize = (10,10))
ax1.plot(data, np.sin(data))
ax2.plot(data, np.cos(data))
ax3.plot(data, np.tan(data))
ax4.plot(data, np.exp(data))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots (nrows = 2, ncols = 2, figsize = (10,10))
ax1.scatter (data, np.sin(data))
ax2.pie(data)
ax3.plot(data, np.tan(data))
ax4.plot(data, np.exp(data))
```

```python
fig, ((ax1, ax2, ax5),
      (ax3, ax4, ax6)) = plt.subplots(nrows = 2, ncols = 3, figsize = (10,10))
ax1.scatter(data, np.sin(data))
ax2.pie(data)
ax3.plot(data, np.tan(data))
ax4.plot(data, np.exp(data))
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 3, figsize = (10,10))
ax[0, 0].plot(data)
ax[0, 1].scatter(data, np.sin(data))
```

```python
fig, ax = plt.subplots (figsize = (10,7))
scatter = ax.scatter (df['age'], df['chol'], c = df['target'])
ax.set_xlabel('age')
ax.set_ylabel('cholestoral')
ax.set_title('age and cholestoral')
ax.legend(*scatter.legend_elements(), title = 'target')
plt.axhline(y = np.mean(df['chol']), color = 'r', linestyle = '--')
```
